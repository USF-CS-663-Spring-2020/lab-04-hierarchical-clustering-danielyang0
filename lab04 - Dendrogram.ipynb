# US Census data

```python
import pandas as pd

# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
## https://github.com/dbrizan/CS663-hierarchical-data
census_df = pd.read_csv('USCensus1990-2000rows.data.txt')
census_df = census_df.iloc[:, 1:] # remove case Id
census_df.head()
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

cencus_Z = linkage(census_df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(cencus_Z)
plt.show()
```

# NIPS conference papers

```python
import pandas as pd

# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
## https://github.com/dbrizan/CS663-hierarchical-data
nips_df = pd.read_csv('NIPS_1987-2015.csv')
nips_df = nips_df.iloc[:, 1:] # remove column 0
nips_df.head()
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

nips_Z = linkage(nips_df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(nips_Z)
plt.show()
```
